## Processing with Time and Date column

```python
def hour_to_day_category(time):
    time = int(str(time)[:2])
    
    if time >= 0 and time < 6:
        return 'Dawn'
    if time >= 6 and time < 12:
        return 'Morning'
    if time >= 12 and time < 18:
        return 'Evening'
    if time >= 18 and time <= 23:
        return 'Night'

df['Time'] = df['Time'].apply(lambda time: hour_to_day_category(time)).astype(str)
```

```python
def date_to_week_day(date):
    day_number = date.weekday()
    days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    day_string = days[day_number]
    return day_string

def date_to_month(date):
    month_number = date.month
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul','Aug','Sept','Oct','Nov','Dec']
    month_string = months[month_number-1]
    return month_string


df['week_day'] = df['Date'].apply(lambda date: date_to_week_day(date))
df['month']  = df['Date'].apply(lambda date: date_to_month(date))
```

Now let's just do some explatory analysis with Python. This is a quick way of showing the total sales in both quantity of products and amount of money.

```python
# group the data by data, then find out the sum of everything in 'quantity'
daily_sales = df.groupby('Date')[' Sold_Quantity'].sum()
daily_retail = df.groupby('Date')['Unit_Price_In_LaKh'].sum()
dates = ['01-02-2021', '01-03-2021', '01-04-2021']

# now show the daily sales for the first few days
for date in dates:
    
    # convert the date string to datetime format
    date_obj = pd.to_datetime(date, format='%m-%d-%Y')
    
    # get the weekday name from the date
    weekday_name = date_obj.strftime('%A')
    
    print(f"On {date}, {weekday_name} we sold {daily_sales[date]} for a retail of {daily_retail[date]}")
```

```python

# group the dataframe by 'weekday' and sum the values in 'quantity'
weekday_totals = df.groupby('week_day')[' Sold_Quantity'].sum()

# print the weekday totals
print(weekday_totals)

# Create a line chart of the weekday_totals
weekday_totals.plot(kind='bar')

# Set the x-axis label and title
plt.xlabel('Day of the Week')
plt.title('Total Quantity Sold by Day of the Week')

# Show the plot
plt.show()
```

So yes, Sunday is the busiest day of the week by far, with Thursday being the slowest day.

```python
# Group the data by article and sum the quantity sold for each article
article_totals = df.groupby('Product_Name')[' Sold_Quantity'].sum()

# Sort the articles by total quantity sold
sorted_articles = article_totals.sort_values(ascending=False)

# Print the top-selling article
print('The best-selling items are:', sorted_articles.head(10))
```

```python
# Group the data by date and article, and sum up the quantity for each group
daily_sales = df.groupby(['Date', 'Product_Name'])[[' Sold_Quantity', 'Unit_Price_In_LaKh']].sum()
# slice that shows how much we made in a day
total_sales = df.groupby('Date')['Unit_Price_In_LaKh'].sum()
# now the total sales for each week
total_sales_weekly = df.groupby(pd.Grouper(freq='W', key='Date'))['Unit_Price_In_LaKh'].sum()

# Show the resulting dataframes
print(daily_sales.head(10))
print(total_sales.head(10))
print(total_sales_weekly.head)
```

```python
# Grouping the data by year and finding the top 2 days for year
top_2_days_2021 = total_sales_weekly['2021'].nlargest(2)


# Plotting the weekly sales for both years
plt.plot(total_sales_weekly['2021'].index, total_sales_weekly['2021'].values, label='2021')


# Adding labels and title
plt.xlabel('Week')
plt.ylabel('Total Sales')
plt.title('Comparison of Weekly Sales by Year',pad=30)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding labeled points to the top 2 days for each year
for week, sales in top_2_days_2021.iteritems():
    plt.annotate(f'{week}\n{sales:.2f}', xy=(week, sales), xytext=(week, sales + 100),
                 ha='left', color='red', fontsize=10)


# Adding a legend
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))

# Display the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor, RandomForestRegressor,BaggingRegressor


df_month_dummy = LabelEncoder().fit_transform(df['month'])
df_item_dummy =LabelEncoder().fit_transform(df['Product_Name'])


X = pd.DataFrame()
X['month'] = df_month_dummy
X['item'] = df_item_dummy
X['week_day'] = df_weekday_dummy


y = df[' Sold_Quantity']

df_month_dummy = LabelEncoder().fit_transform(df['month'])
df_item_dummy =LabelEncoder().fit_transform(df['Product_Name'])
df_weekday_dummy = LabelEncoder().fit_transform(df['week_day'])
df_weekday_dummy = LabelEncoder().fit_transform(df['week_day'])
X = pd.DataFrame()
X['month'] = df_month_dummy
X['item'] = df_item_dummy
X['week_day'] = df_weekday_dummy
y = df[' Sold_Quantity']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)

def run_model(model,X_train,X_test,y_train,y_test):
    model = model
    model.fit(X_train,y_train)
    y_predictions = model.predict(X_test)
    df_result = pd.DataFrame()
    df_result['Predictions'] = np.round(y_predictions)
    df_result['True Value'] = y_test.values
    plt.figure(figsize=(12,4),dpi= 300)
    sns.set(style='dark')
    sns.lineplot(data=df_result,x = df['Date'], y= 'True Value')
    sns.lineplot(data=df_result,x = df['Date'], y= 'Predictions')
    print(df_result.describe())
    print('MSE:'+ str(mean_squared_error(y_test,y_predictions)))
    print('RMSE:'+ str(np.sqrt(mean_squared_error(y_test,y_predictions))))
    print('MASE:'+ str(mean_absolute_error(y_test,y_predictions)))
    
    return df_result

def predict_items_today():
    month = datetime.date.today().month
    weekday = datetime.date.today().weekday()

    resultPrediction =[]
    itemName =[]
    for index,item in enumerate(df['Product_Name'].unique()):
        test = [(month,index,weekday),]
        resultPrediction.append(int(np.round(model.predict(test))))
        itemName.append(item)
        
    newDf = pd.DataFrame()
    newDf['item_name'] = itemName
    newDf['prediction'] = resultPrediction
    print("Item demand today: " + (str(newDf['prediction'].sum())) + " items")
    print("Total Gains Today: " + (str(int(np.round(( df['Unit_Price_In_LaKh'] * newDf['prediction']).sum()))) + 'Lakh'))
    return newDf


def predict_items_by_date(date):
    if not isinstance(date, datetime.date):
        raise ValueError("Invalid date format. Please provide a valid date object.")

    month = date.month
    weekday = date.weekday()
    resultPrediction =[]
    itemName =[]
    for index, item in enumerate(df['Product_Name'].unique()):
        test = [(month, index, weekday),]
        resultPrediction.append(int(np.round(model.predict(test))))
        itemName.append(item)
        
    newDf = pd.DataFrame()
    newDf['item_name'] = itemName
    newDf['prediction'] = resultPrediction
    print("Item demand on {}: {}".format(date.strftime("%Y-%m-%d"), str(newDf['prediction'].sum())) + " items")
    print("Predicted Gains on {}: {}Lakh".format(date.strftime("%Y-%m-%d"), str(int(np.round((df['Unit_Price_In_LaKh'] * newDf['prediction']).sum())))))
    return newDf



predict_items_today()
predict_items_by_date(datetime.date(2022, 2, 1))
```
